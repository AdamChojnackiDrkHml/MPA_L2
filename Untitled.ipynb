```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dfNorm = pd.read_csv("test/testNorm.csv", delimiter=";")
dfUni = pd.read_csv("test/testUni.csv", delimiter=";")
```

```python
def getFromColumn(column):
    return column.min(), column.max(), column.mean(), column.std(), column.var()

def extractFromColumnByN(df, columnName, ns):
    minC = []
    maxC = []
    meanC = []
    stdC = []
    varC = []
    for n in ns:
        column = df.loc[df.n == n][columnName]
        minC.append(column.min())
        maxC.append(column.max())
        meanC.append(column.mean())
        stdC.append(column.std())
        varC.append(column.var())
        
    return minC, maxC, meanC, stdC, varC
```

```python
ns = dfNorm.n.unique()

minKruskalN, maxKruskalN, meanKruskalN, stdKruskalN, varKruskalN = extractFromColumnByN(dfNorm, "kruskal", ns)
minPrimN, maxPrimN, meanPrimN, stdPrimN, varPrimN = extractFromColumnByN(dfNorm, "prim", ns)
minWeightN, maxWeightN, meanWeightN, stdWeightN, varWeightN = getFromColumn(dfNorm.mst.values)

minKruskalU, maxKruskalU, meanKruskalU, stdKruskalU, varKruskalU = extractFromColumnByN(dfUni, "kruskal", ns)
minPrimU, maxPrimU, meanPrimU, stdPrimU, varPrimU = extractFromColumnByN(dfUni, "prim", ns)
minWeightU, maxWeightU, meanWeightU, stdWeightU, varWeightU = getFromColumn(dfUni.mst.values)
```

```python
avgWeightByNUni = []
avgWeightByNNorm = []

for n in ns:
    weightsUni = dfUni.loc[dfUni.n == n].mst
    weightsNorm = dfNorm.loc[dfNorm.n == n].mst
    
    avgWeightByNUni.append(weightsUni.mean())
    avgWeightByNNorm.append(weightsNorm.mean())

fig, axs = plt.subplots(1, 2,  figsize=(16, 6))

axs[0].plot(ns, avgWeightByNNorm)
axs[0].set_xlabel("N")
axs[0].set_ylabel("Mst weight")
axs[0].set_title("Średni rozmiar MST dla rozkładu normalnego")
axs[1].plot(ns, avgWeightByNUni)
axs[1].set_xlabel("N")
axs[1].set_ylabel("Mst weight")
axs[1].set_title("Średni rozmiar MST dla rozkładu jednostajnego")
plt.show()
plt.clf()
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
_ = sns.kdeplot(dfNorm.mst.values, bw_method=0.5, ax=axs[0]).set(title='Rozkład wag MST dla rozkładu normalnego')
_ = sns.kdeplot(dfUni.mst.values, bw_method=0.5, ax=axs[1]).set(title='Rozkład wag MST dla rozkładu jednostajnego')
```

```python
def printRange(minT, meanT, maxT, name, ax, dist):
    ax.plot(ns, minT, label="min")
    ax.plot(ns, meanT, label="mean")
    ax.plot(ns, maxT, label="max")
    ax.set_xlabel("N")
    ax.set_ylabel("time [s]")
    ax.legend()
    ax.set_title(f"Czas działania algorytmu {name} rozkład {dist}")


fig, axs = plt.subplots(2, 2,  figsize=(12, 8))
printRange(minKruskalN, meanKruskalN, maxKruskalN, "Kruskal", axs[0][0], "normalny")
printRange(minPrimN, meanPrimN, maxPrimN, "Prim", axs[0][1], "normalny")
printRange(minKruskalU, meanKruskalU, maxKruskalU, "Kruskal", axs[1][0], "jednostajny")
printRange(minPrimU, meanPrimU, maxPrimU, "Prim", axs[1][1], "jednostajny")
fig.tight_layout()
# printRange(minWeight, meanWeight, maxWeight)
```
