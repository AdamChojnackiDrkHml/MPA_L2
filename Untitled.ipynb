```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import math
```

```python
dfNorm = pd.read_csv("data/resultTestNorm.csv", delimiter=";")
dfUni = pd.read_csv("data/resultTestUni.csv", delimiter=";")
```

```python
def getFromColumn(column):
    return column.min(), column.max(), column.mean(), column.std(), column.var()

def extractFromColumnByN(df, columnName, ns):
    minC = []
    maxC = []
    meanC = []
    stdC = []
    varC = []
    for n in ns:
        column = df.loc[df.n == n][columnName]
        minC.append(column.min())
        maxC.append(column.max())
        meanC.append(column.mean())
        stdC.append(column.std())
        varC.append(column.var())
        
    return minC, maxC, meanC, stdC, varC
```

```python
ns = dfNorm.n.unique()

minKruskalN, maxKruskalN, meanKruskalN, stdKruskalN, varKruskalN = extractFromColumnByN(dfNorm, "kruskal", ns)
minPrimN, maxPrimN, meanPrimN, stdPrimN, varPrimN = extractFromColumnByN(dfNorm, "prim", ns)
minWeightN, maxWeightN, meanWeightN, stdWeightN, varWeightN = getFromColumn(dfNorm.mst.values)

minKruskalU, maxKruskalU, meanKruskalU, stdKruskalU, varKruskalU = extractFromColumnByN(dfUni, "kruskal", ns)
minPrimU, maxPrimU, meanPrimU, stdPrimU, varPrimU = extractFromColumnByN(dfUni, "prim", ns)
minWeightU, maxWeightU, meanWeightU, stdWeightU, varWeightU = getFromColumn(dfUni.mst.values)
```

```python
avgWeightByNUni = []
avgWeightByNNorm = []

for n in ns:
    weightsUni = dfUni.loc[dfUni.n == n].mst
    weightsNorm = dfNorm.loc[dfNorm.n == n].mst
    
    avgWeightByNUni.append(weightsUni.mean())
    avgWeightByNNorm.append(weightsNorm.mean())

fig, axs = plt.subplots(1, 2,  figsize=(16, 6))

def gamma3(n):
    return sum([1/(i**2) for i in range (1, n, 1)])

axs[0].plot(ns, avgWeightByNNorm)
axs[0].set_xlabel("N")
axs[0].set_ylabel("Mst weight")
axs[0].set_title("redni rozmiar MST dla rozkadu normalnego")
axs[1].plot(ns, avgWeightByNUni)
axs[1].axhline(y = sum(avgWeightByNUni)/len(avgWeightByNNorm), c='r', linestyle='-', label='mean')
axs[1].set_yticks(list(axs[1].get_yticks()) + [sum(avgWeightByNUni)/len(avgWeightByNNorm)])
axs[1].set_xlabel("N")
axs[1].set_ylabel("Mst weight")
axs[1].set_title("redni rozmiar MST dla rozkadu jednostajnego")
plt.show()
plt.clf()
```

```python
def chebyshev(data, percent):
        EX = sum([abs(dat) for dat in data]) / (len(data)- 1)
        VarX = sum([(dat - EX) ** 2 for dat in data]) / (len(data) - 1)
        return math.sqrt(VarX/ percent)

def getUpperAndLower(data, percent):
    sortedData = sorted(data)
    return sortedData[int(percent * len(data))], sortedData[-(int(percent * len(data)) + 1)]
```

```python
percent = 0.2
chebyshevPrimU = [chebyshev(dfUni.loc[dfUni.n == n].prim, percent) for n in ns]
realPrimU = [getUpperAndLower(dfUni.loc[dfUni.n == n].prim, percent) for n in ns]
chebyshevKruskalU = [chebyshev(dfUni.loc[dfUni.n == n].kruskal, percent) for n in ns]
realKruskalU = [getUpperAndLower(dfUni.loc[dfUni.n == n].kruskal, percent) for n in ns]
chebyshevPrimN = [chebyshev(dfNorm.loc[dfNorm.n == n].prim, percent) for n in ns]
realPrimN = [getUpperAndLower(dfNorm.loc[dfNorm.n == n].prim, percent) for n in ns]
chebyshevKruskalN = [chebyshev(dfNorm.loc[dfNorm.n == n].kruskal, percent) for n in ns]
realKruskalN = [getUpperAndLower(dfNorm.loc[dfNorm.n == n].kruskal, percent) for n in ns]

def plotWithChebyshevAndReal(cheby, mean, real, ax, ns, title):
    lowerReal = list(map(lambda x: x[1], real))
    upperReal = list(map(lambda x: x[0], real))
    chebyLow = [mean[i] - cheby[i] for i in range(len(ns))]
    chebyHigh = [mean[i] + cheby[i] for i in range(len(ns))]
    ax.plot(ns, lowerReal, label = "lower real")
    ax.plot(ns, upperReal, label = "upper real")
    ax.plot(ns, chebyLow, label = "lower cheby")
    ax.plot(ns, chebyHigh, label = "upper cheby")
    ax.legend()
    ax.set_title(title)

fig, axs = plt.subplots(4,  figsize=(10, 40))
plotWithChebyshevAndReal(chebyshevPrimU, meanPrimU, realPrimU, axs[0], ns, "Koncentracja czasu dla alg Prima, rozkad jednostajny")
plotWithChebyshevAndReal(chebyshevKruskalU, meanKruskalU, realKruskalU, axs[1], ns, "Koncentracja czasu dla alg Kruskala, rozkad jednostajny")
plotWithChebyshevAndReal(chebyshevPrimN, meanPrimN, realPrimN, axs[2], ns, "Koncentracja czasu dla alg Prima, rozkad normalny")
plotWithChebyshevAndReal(chebyshevKruskalN, meanKruskalN, realKruskalN, axs[3], ns, "Koncentracja czasu dla alg Kruskala, rozkad normalny")


plt.show()
plt.clf()
```
